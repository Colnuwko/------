6. Для матрицы, созданной в 4 пункте, найти LU разложение
самостоятельно написанной функцией и проверить результат с
помощью перемножения матриц.

```python
line_8x8 = np.random.randint(0, 10, (8,8))
def LU(line_8x8):
    size = line_8x8[0].size
    print(size)
    U = np.zeros((size,size), float)
    L = np.identity(size, float)
    for i in range(size):
        L[i, i] = 1
        for k in range(i, size):
            sum_uk = 0
            sum_lk = 0
            for j in range(i):
                sum_uk += L[i, j] * U[j, k]
                sum_lk += L[k, j] * U[j, i]
            U[i, k] = line_8x8[i, k] - sum_uk
            L[k, i] = (line_8x8[k, i] - sum_lk) / U[i, i]
    return L, U
L, U = LU(line_8x8)
np.set_printoptions(precision=2, suppress=True)
print("U = ",U,"\nL = ",L)

```

```python
print(line_8x8)
print("Проверка \n",np.dot(L, U))
```

```python
#алгоритм грамма шмидта
line_8x8 = line_8x8.astype(float)
n = line_8x8.shape[1]
Q = np.zeros_like(line_8x8)
R = np.zeros((n, n))

for j in range(n):
    v = line_8x8[:, j]
    for i in range(j):
        R[i, j] = np.dot(Q[:, i], line_8x8[:, j])
        v = v - (R[i, j] * Q[:, i])
    R[j, j] = np.linalg.norm(v)
    Q[:, j] = v / R[j, j]



print("\nМатрица Q (Грама-Шмидта):\n", Q)
print("\nМатрица R (Грама-Шмидта):\n", R)

```

Метод Гвинеса

```python
def gvines(line_8x8):
    (r, c) = np.shape(line_8x8)
    Q = np.identity(r)
    R = np.copy(line_8x8)
    (rows, cols) = np.tril_indices(r, -1, c)
    for (row, col) in zip(rows, cols):
        if R[row, col] != 0: # Q = 1, S = 0, я, Q без изменений
            r_ = np.hypot(R[col, col], R[row, col]) # d
            c = R[col, col]/r_
            s = -R[row, col]/r_
            G = np.identity(r)
            G[[col, row], [col, row]] = c
            G[row, col] = s
            G[col, row] = -s
            R = np.dot(G, R) # R=G(n-1,n)*…*G(2n)*…*G(23,1 n)*…*G(12)*A
            Q = np.dot(Q, G.T) # *Q-G(n-1,n).7*…*G(2 n).7*…**G(23,in).7*…*G(12).7
    return Q, R
Q, R = gvines(line_8x8)
np.set_printoptions(precision=4, suppress=True)
print(Q)
print(R)
```

. Решить систему методом простых итераций с точностью до
10−3
. Проверить выполнение достаточного условия сходимости.
Если условие не выполняется, в программе выполнить эквивалентные преобразования системы, после этого привести к удобному
для итераций виду. Оформить итерации в виде таблицы (можно
пользоваться модулями PrettyTable или Pandas).

{
3.2x1 −2.5x2 +3.7x3 = 6.5
0.5x1 +0.34x2 +1.7x3 = −0.24
1.6x1 +2.3x2 −1.5x3 = 4.3
}

```python
from prettytable import PrettyTable
A = np.array([[3.2, -2.5, 3.7], [0.5, 0.34, 1.7], [1.6, 2.3, -1.5]])
B = np.array([6.5, -0.24, 4.3])
n = len(B)
print("Исходная матрица: \n", A)
print(B)
#проверка сходимости
def check(matrix):
    n = len(matrix)
    for i in range(n):
        tmp_sum = 0
        for j in range(n):
            if (j != i):
                tmp_sum += abs(matrix[i, j])
        if tmp_sum > abs(matrix[i, i]):
            print("Достаточное условие сходимости не выполняется")
            return
    print("Достаточное условие сходимости выполняется")
    return

check(A)
#у исходной матрицы условие сходимости не выполняется, значит требуется преобразовать матрицу к нужному виду
#переставляем строки
A = np.array([[4.1, 5.8, -1.7], [2.7, 3.3, 1.3], [3.5, -1.7, 2.8]])
B = np.array([0.8, 2.1, 1.7])

#складываем 1-ую и 3-ю строки
A[0,0] += A[2,0]
A[0,1] += A[2,1]
A[0,2] += A[2,2]
B[0] += B[2]
# print(A)
# print(B)

#вычитаем из 2-ой , 3-ю строку
A[1,0] -= A[2,0]
A[1,1] -= A[2,1]
A[1,2] -= A[2,2]
B[1] -= B[2]
# print(A)
# print(B)

#3-ю строку умножаем на 2
A[2,0] *= 2
A[2,1] *= 2
A[2,2] *= 2
B[2] *= 2
# print(A)
# print(B)

#из 3-ей строки вычитаем 1-ую строку умноженную на 0.5
A[2,0] -= A[0,0]*0.5
A[2,1] -= A[0,1]*0.5
A[2,2] -= A[0,2]*0.5
B[2] -= B[0]*0.5
# print(A)
# print(B)

#складываем 3-ю и 2-ую строки
A[2,0] += A[1,0]
A[2,1] += A[1,1]
A[2,2] += A[1,2]
B[2] += B[1]
print(A)
print(B)

check(A)

# Инициализация начального приближения x(0)
x = np.zeros_like(B)

# Заданная точность
tolerance = 1e-3

# Максимальное количество итераций (чтобы избежать бесконечного цикла)
max_iterations = 1000

# Создадим таблицу для вывода итераций
table = PrettyTable()
table.field_names = ["k", "x_1", "x_2", "x_3", "||x(k+1) - x(k)||"]

D = np.diag(np.diag(A))
F = A - D

for k in range(max_iterations):
    # Рассчитываем x(k+1)
    x_new = np.linalg.inv(D).dot(B - F.dot(x))

    # Вычисляем норму разницы между x(k+1) и x(k)
    norm = np.linalg.norm(x_new - x)

    # Добавляем текущую итерацию в таблицу
    table.add_row([k, x[0],x[1],x[2], norm])

    # Обновляем x для следующей итерации
    x = x_new

    # Проверяем условие завершения итераций
    if norm < tolerance:
        break

# Выводим результаты итераций
print(table)

# Выводим итоговое решение
print("Итоговое решение (x): ", x)

print("linalg.solve(): ", np.linalg.solve(A, B))
```
